```python
from bokeh.plotting import figure , output_file,show
```

```python
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

```python
output_file('test.html')
p=figure(title='test flower')
p.xaxis.axis_label='petal _length'
p.yaxis.axis_label = 'petal_width'
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
output_file('test.html')
p=figure(title='test flower')
p.xaxis.axis_label='petal _length'
p.yaxis.axis_label = 'petal_width'
p.line(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='line graph')
p.line(x,y)
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='line graph')
p.scatter(x,y)
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='line graph')
p.scatter(x,y,size=20,fill_color='green')
show(p)
```

```python
x=[3,4,5,6,7]
y=[4,5,6,7,4]
output_file('line.html')
p=figure(title='line graph')
p.scatter(x,y,size=20,fill_color='green',legend_label='This is my data point')
show(p)
```
